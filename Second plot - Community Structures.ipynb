```r
library(igraph)
setwd('C:/Users/Yuting/Downloads/project2')
filenames = list.files("gplus")
```

```r
# Extract the community structure of each personal network using Walktrap community detection algorithm.
edges_file = paste("gplus/","109213135085178239952" , ".edges", sep="")
g1_network = read.graph(edges_file , format = "ncol" , directed=TRUE)  
node_size = rep(2, vcount(g1_network))
walktrap_comm1 = walktrap.community(g1_network)

plot(walktrap_comm1,g1_network, asp = 9/16,vertex.label=NA , edge.color = "grey",vertex.size=node_size,main=paste("Walktrap Community of network", 1))
```

```r
edges_file = paste("gplus/","115625564993990145546" , ".edges", sep="")
g2_network = read.graph(edges_file , format = "ncol" , directed=TRUE)  
node_size = rep(2, vcount(g2_network))
walktrap_comm2 = walktrap.community(g2_network)

plot(walktrap_comm2,g2_network, asp = 9/16,vertex.label=NA , edge.color = "grey",vertex.size=node_size,main=paste("Walktrap Community of network", 2))
```

```r
edges_file = paste("gplus/" , "101373961279443806744"  , ".edges" , sep="")
g3_network = read.graph(edges_file , format = "ncol" , directed=TRUE)  
node_size = rep(2, vcount(g3_network))
walktrap_comm3 = walktrap.community(g3_network)
cat("modularity scores of 3rd personal network: ", modularity(walktrap_comm3),"\n")


plot(walktrap_comm3,g3_network, asp = 9/16,vertex.label=NA , edge.color = "grey",vertex.size=node_size,main=paste("Walktrap Community of network", 3))
```

As we can seee from three networks, the user points belonging to different communities are not well seperated, and all the "outliers" are considered as single communities. We can further test the precision of walktrap algorithm on Google+ dataset by comparing the communities to ground-truth circles. By calculating the homogeneity and completeness, and test on more nodes, we can summrize under what conditions the walktrap algorithm would be most accurate.
