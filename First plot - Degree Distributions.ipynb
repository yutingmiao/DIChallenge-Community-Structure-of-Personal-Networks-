```r
# Compare the in-degree and out-degree distribution of 4 personal networks.
par(mfrow = c(3,2))
nodes =c("100329698645326486178","109213135085178239952","115625564993990145546")
for (i in 1:3) {
    edges_file = paste("gplus/", nodes[i] , ".edges", sep="")
    gnetwork = read.graph(edges_file, format = "ncol", directed=TRUE)  
    
    degree_in <- degree.distribution(gnetwork, mode="in")
    plot(degree_in, xlab="Degree", ylab="Frequency", main=paste("In-degree distribution of network", i))
    d1_out<-degree.distribution(gnetwork,mode="out")
    plot(d1_out, xlab="degree", ylab="frequency", main=paste("Out-degree distribution of netwrok", i))
}
```

As we can see from the distributions of 3 different personal nodes, some in-degree distributions are close to the power law distribution, and most of the nodes are below degree 100, meaning that most users have fewer than 100 connections. However, the maximum frequencies vary in differnet networks. Out-degree distributions look similar for different networks. They all have the highest frequency at degree 1 and most of the nodes are below degree 10.
